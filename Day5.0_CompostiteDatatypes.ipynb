### Tuple

*A tuple is similar to a List, but it's immutable.*
*It is a read-only type of list.*
*We can enclose the elements of a tuple in ( ).*

```python
t=(10,20,30,40,50)
type(t)
```

```python
# t.append(22) ->error - item assignment doen't work in tuple
```

```python
t[0]
```

*It represents a sequnce of numbers.*
*The elements present in range a cannot be modified.*
*It is immutable.*

```python
a=range(10)
a
```

```python
for i in a: print(i)
```

```python
b=range(11,20)
```

```python
for i in b: print(i)
```

### Set Data type

*When we want to represent a group of values without duplicates.*
*Where order is not important, then we should use the set data type.*

###### 1. *Insertion order is not preserved*
###### 2. *Duplicates are not allowed*
###### 3. *Hetrogeneous objects are allowed*
###### 4. *Index concept is not applicable in set*
###### 5. *It is mutable*
###### 6. *Growable in nature*
###### 7. *Enclosed in { }*

```python
s={12,23.34,3+8j,"Cole",True,12}
```

```python
s # duplicates will not be repeated
```

```python
#s[1] -> error --- indexing is not allowed
```

```python
s.add(56)
```

```python
s
```

```python
s.remove(12)
```

```python
s
```

### Frozenset Data type

*A frozenset is similar to a set, but it is immutable. Methods like add(), remove() will not work.*

```python
s
```

```python
fs=frozenset(s)
```

```python
type(fs)
```

```python
for i in fs: print(i)
```

```python
# fs.add(44) & fs.remove(56) -> will not work
```

### Dict Data type

*If we want to represent a group of values in Key-Value pair form. Growable in nature -> can be modified*

```python
d={101:'mango',101:'banana',103:'kiwi'}
d
```

```python
food_item={"Starter":"Chicken65","Mains":"Ika Mata","Dessert":"Crepes"}
food_item
```

**Key : Value**
1. If key is number it will not be in quotations
2. If it is String then key should be in quatations
3. Value will always be in quotations

***Note -> Keys cannot be the same. If duplicate keys are used, the program will run, but only the last value will be stored. However, duplicate values are allowed.***

```python
food_item["Mains"]="Coxinhas"
```

```python
food_item
```

```python
#blank dictionary 
bd={}
type(bd)
```
