### Iterative Statements

*If we want to execute a group of statements multiple times, then we have to go for iterative statements.*
1. For
2. While

#### for loop

*If we want to execute some action for every element present in a sequence(collection of string or others),then we should go for - for loop*

- *sequence - a string or any collection.*
- *body will be executed on each element present in string/collection.*

```python
s='nikolai sokolov'
for x in s:
    print(x)
```

```python
list1={10,34.5,23,42,'name'}
for x in list1:
    print(x)
```

```python
# To print character present in string-indexwise
name=input("Enter a name")
i=0
for x in name:
    print("Character present at",i,"index is:",x)
    i+=1
```

```python
# print 'hello' 10 times
a="hello"
for i in range(10):
    print(a)
```

```python
for i in range(10):
    print("hello")
```

```python
# table of 2
for i in range(2,22,2):
    print(i)
```

```python
# print from 10 to 1
for i in range(10,0,-1):
    print(i)
```

```python
# print odd nos. from 1 to 20
for i in range(1,20,2):
    print(i)
```

```python
# print even nos. from 1 to 20
for i in range(2,21,2):
    print(i)
```

```python
for x in range(21):
    if(x%2!=0):
        print(x)
```

```python
# Take number as input and create the table
a=int(input("Enter a value"))
for i in range(a,a*11,a):
    print(i)    
```

#### while loop

*If we want to execute a group of statements iteratively until the condition becomes false, then we should go for while loop.*

```python
# to print nos. from 1 to 10
x=1
while x<=10:
    print(x)
    x+=1
```

```python
#display the sum of first n number
n=int(input("Enter a num"))
sum=0
i=1
while i<=n:
    sum+=i
    i+=1
    print("the sum of first",n,"number is",sum)
```

#### nested loop

*loop within the loop*

```python
for i in range(4):
    for j in range(4):
        print("i=",i,"j=",j)
```

### Transfer Statements

#### break

*We can use break statement inside loops to break the loop execution based on some condition.*

```python
for i in range(10):
    if i==7:
        print("Processing is enough, please break")
        break
    print(i)
```

```python
# using break in list
cart=[100,200,300,400,500,600]
for item in cart:
    if item>500:
        print("you get 20% discount on this order")
        break
    print(item)
```

#### continue

*we can use continue statement to skip current iteration and continue to next.*

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# using continue in list
cart=[122,322,422,300,500,600,700,100,288]
for item in cart:
    if item>=500:
        print("we cannot process this item",item)
        continue
    print(item)
```

#### pass

- *pass is a keyword in python*
- *In python if a block is required which won't do anything then we can define that empty block with pass keyword*

```python
for i in range(100):
    if i%9==0:
        print(i)
    else:
        pass
```
