### Removing spaces from the string

There are 3 methods to remove space:
1. **rstrip()** -> *remove space right hand side*
2. **lstrip()** -> *remove space left hand side*
3. **strip()** -> *remove space both sides*

```python
city=input("Enter your city name")
scity=city.lstrip()
if scity=="Hyderabad":
    print("Hello Hyderabadi...adaab")
elif scity=="Chennai":
    print("Hello Madrasi...vanakam")
elif scity=='Banglore':
    print("Hello subhodaya")
else:
    print("Enter valid city name")
```

### Finding substring

We can find out the substring based on direction.
1. Forward Direction
    - **find()** ->*Return first occurence of substring*
    - **index()** ->*Return the substring*
2. Backward Direction
   - **rfind()** ->*Reverse find*
   - **rindex()** ->*Reverse index*

```python
# using find()
s="Machine learning is learning machine"
s.find('machine')
```

```python
s.find('a',3,36)
```

```python
# using index()
s.index('r',4,40)
```

```python
# using rfind()
s.rfind('r')
```
