# Operators

***Operators are symbols that perform operations.***

There are many types of Operators:
1. *Arithmetic Operators*
2. *Relational / Comparision Operators*
3. *Logical Operators*
4. *Bitwise Operators*
5. *Assignment Operators*
6. *Special Operators*

### Arithmetic Operators

- *Addition* -> +
- *Subtraction* -> -
- *Multiplication* -> *
- *Division* -> /
- *Modulo* -> %
- *Floor Division* -> //
- *Exponential or Power* -> **

```python
a=10
b=2
print("Addition->",a+b)
print("Subtraction->",a-b)
print("Multiplication->",a*b)
print("Division->",a/b)
print("Modulo->",a%b)
print("Floor Division->",a//b)
print("Power->",a**b)
```

```python
# a/0 -> Zerodivision error
```

### Relational Operators

- *Greater than(>)*
- *Greater than or equal to(>=)*
- *Less than(<)*
- *Less than or equal to(<=)*

```python
a=50
b=25
print("a>b?",a>b)
print("a>=b?",a>=b)
print("a<b?",a<b)
print("a<=b?",a<=b)
```

#### Equality Operators

- *Not Equals to(!=)*

### Bitwise Operator

*These operators can only be used with integer or boolean datatypes*

Conditions:
- Bitwise AND **`&`** -> If both bits are 1, result will be 1 - otherwise 0
- Bitwise OR **`|`** -> If at least one bit is 1, result will be 1 - otherwise 0
- Bitwise XOR **`^`** -> If bits are different, result will be 1 -otherwise 0
- Bitwise left shift **`<<`** -> Shift bits to the left
- Bitwise right shift **`>>`** -> Shift bits to the right

```python
bin(4)
```

```python
bin(5)
```

```python
print(4&5)
```

```python
bin(10)
```

```python
bin(15)
```

```python
print(10|15)
```

```python
print(10^15)
```

```python
bin(10)
```

```python
bin(2)
```

```python
print(10<<2)
```

```python
print(10>>2)
```
