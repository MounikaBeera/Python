```python
# new line-> \n
da="Data\nAnalytics"
print(da)
```

In String literals, we can use escape characters to associate a spl meaning***

```python
# Tab distance-> \t
wd="Web\tDevelopment"
print(wd)
```

```python
# Backspace-> \b
gd="Graphic\bDesign"
print(gd)
```

```python
# Vertical tab-> \v
jd="Job\vDescription"
print(jd)
```
